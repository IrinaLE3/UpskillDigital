```python
# create a board
board = ["-", "-", "-",
         "-", "-", "-",
         "-", "-", "-"]

# display a board
def show_board():
    print(board[0] + "|" + board[1] + "|" + board[2])
    print(board[3] + "|" + board[4] + "|" + board[5])
    print(board[6] + "|" + board[7] + "|" + board[8])

# functions for the main program    

def ask_player1():
    box_number = int(input("Player1: Please enter cell number from 0 to 8: "))
    if board[box_number] == "-":
        board[box_number] = "X"
    else:
        print("Incorrect move")

def ask_player2():
    box_number = int(input("Player2: Please enter cell number from 0 to 8: "))
    if board[box_number] == "-":
        board[box_number] = "0"
    else:
        print("Incorrect move")
    

def check_victory1():
    if board[0] == "X" and board[1] =="X" and board[2] == "X":
        return True
    if board[3] == "X" and board[4] =="X" and board[5] == "X":
        return True
    if board[6] == "X" and board[7] =="X" and board[8] == "X":
        return True
    
    if board[0] == "X" and board[3] =="X" and board[6] == "X":
        return True
    if board[1] == "X" and board[4] =="X" and board[7] == "X":
        return True
    if board[2] == "X" and board[5] =="X" and board[8] == "X":
        return True
    
    if board[0] == "X" and board[4] =="X" and board[8] == "X":
        return True
    if board[6] == "X" and board[4] =="X" and board[2] == "X":
        return True
    
    return False

def check_victory2():
    if board[0] == "0" and board[1] =="0" and board[2] == "0":
        return True
    if board[3] == "0" and board[4] =="0" and board[5] == "0":
        return True
    if board[6] == "0" and board[7] =="0" and board[8] == "0":
        return True
    
    if board[0] == "0" and board[3] =="0" and board[6] == "0":
        return True
    if board[1] == "0" and board[4] =="0" and board[7] == "0":
        return True
    if board[2] == "0" and board[5] =="0" and board[8] == "0":
        return True
    
    if board[0] == "0" and board[4] =="0" and board[8] == "0":
        return True
    if board[6] == "0" and board[4] =="0" and board[2] == "0":
        return True
    
    return False

# start a game    

show_board()                           # displays initial board

count = 0                              # initialising count variable

for i in range(5):
    
    ask_player1()                       # calling ask_player1 function
    count = count + 1                   # counting moves ( to tell if it's a tie or not)                    
    show_board()                        # displays board with changes made by Player1
    player1_victory = check_victory1()  # checking if Player1 has won
    if player1_victory == True:
        print("Congratulations! Player 1 has won!")
        break  
    if count == 9:
        print("It is a tie")
        break                            # loop exit
        
    ask_player2()
    count = count + 1
    show_board()
    player2_victory = check_victory2()
    if player2_victory == True:
        print("Congratulations! Player 2 has won!")
        break
    if count == 9:
        print("It is a tie")
        break    
    
```
