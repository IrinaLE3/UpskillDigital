```python
# replace empty cells with NaN:
data = data.replace('', np.nan)
```

```python
data["Coarse wool Price"] = data ["Coarse wool Price"].str.replace (',','')
```

```python
data["Fine wool Price"] = data ["Fine wool Price"].str.replace (',','')
```

```python
data["Copra Price"] = data ["Copra Price"].str.replace (',','')
```

```python
data = data.replace('-', '')
```

```python
data.dtypes
```

```python
# Drop rows with NaN values:
data = data.dropna()
```

```python
# Check if all NaN values have been deleted:
data.isnull().sum()
```

```python
# Delete unnecessary columns:
data.drop(["Coarse wool price % Change","Copra price % Change","Cotton price % Change","Fine wool price % Change","Hard log price % Change","Hard sawnwood price % Change","Hide price % change","Plywood price % Change","Rubber price % Change","Softlog price % Change","Soft sawnwood price % Change","Wood pulp price % Change"], axis = 1,inplace = True)
```

```python
# Prepare a list of columns to convert object into float:
col_to_convert = ["Coarse wool Price","Copra Price","Fine wool Price"]
```

```python
# Convert object into float:
data[col_to_convert] = data[col_to_convert].astype("float")
```

```python
data.dtypes
```

```python
data.head()
```

```python
data.Month  = pd.to_datetime(data["Month"].str.upper(), format='%b-%y', yearfirst=False)

# Indexing month
data = data.set_index('Month')
```

```python
data.plot(
    y= ['Coarse wool Price','Copra Price','Cotton Price','Fine wool Price','Hard log Price','Hard sawnwood Price','Hide Price','Plywood Price','Rubber Price','Softlog Price','Soft sawnwood Price','Wood pulp Price']
   ,figsize=(16, 8),title = "Raw materials price change over the past 17 years")
```

```python
a = data.loc["1990-04-01", "Coarse wool Price"]
b = data.loc["1990-04-01","Fine wool Price"]
c = np.array([a, b])
mylabels = ["Coarse wool price", "Fine wool price"]
plt.pie(c, labels = mylabels)
plt.title("Coarse and fine wool prices in 1990")
plt.show()

a = data.loc["2016-04-01", "Coarse wool Price"]
b = data.loc["2016-04-01","Fine wool Price"]
c = np.array([a, b])
mylabels = ["Coarse wool price", "Fine wool price"]
plt.pie(c, labels = mylabels)
plt.title("Coarse and fine wool prices in 2016")
plt.show()
```

```python
a = data.loc["1990-04-01", "Plywood Price"]
b = data.loc["2016-04-01","Plywood Price"]
plywood_growth = (b - a)/ a * 100
plywood_growth
d = data.loc["1990-04-01", "Soft sawnwood Price"]
e = data.loc["2016-04-01","Soft sawnwood Price"]
soft_sawnwood_price_growth = (e - d)/ d * 100
soft_sawnwood_price_growth

x = np.array(["Plywood","Soft sawnwood"])
y = np.array([plywood_growth,soft_sawnwood_price_growth])
plt.bar(x,y)
plt.title("Price percentage difference between 1990 and 2016")
plt.show()
```
